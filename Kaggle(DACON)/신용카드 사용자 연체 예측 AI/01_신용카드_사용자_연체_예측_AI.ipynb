- Train/Test를 한꺼번에 처리하기 위해 concat

```python
data = pd.concat([train, test], axis = 0)
data.shape
```

```python
data = data.reset_index()
data
```

```python
data.isnull().sum()
```

- 일일이 분석하기 힘든 직업은 제거

```python
data = data.drop('occyp_type', axis = 1)
```

```python
data.isnull().sum()
```

['Y', 'N'], ['F', 'M']과 같이 계산할 수 없는 경우 0, 1 로 map 시켜준다.

- 성별

```python
data['gender'] = data['gender'].replace(['F', 'M'], [0, 1])
print(data['gender'].value_counts())
```

- 차량 유/무

```python
data['car'] = data['car'].replace(['N', 'Y'], [0, 1])
print(data['car'].value_counts())
```

- 부동산 소유 유/무

```python
data['reality'] = data['reality'].replace(['N', 'Y'], [0, 1])
print(data['reality'].value_counts())
```

- 구체적인 수치로 표현하기 어려우므로 따로 판단하기 위해서 one-hot encoding 적용

```python
print(data['income_type'].unique())
print(data['edu_type'].unique())
print(data['family_type'].unique())
print(data['house_type'].unique()) 
```

```python
F1 = pd.get_dummies(data.income_type, prefix = 'income_type')
F2 = pd.get_dummies(data.edu_type, prefix = 'edu_type')
F3 = pd.get_dummies(data.family_type, prefix = 'family_type')
F4 = pd.get_dummies(data.house_type, prefix = 'house_type')

F1
```

```python
data = pd.merge(data, F1, left_index = True, right_index = True, how = 'left' )
data = pd.merge(data, F2, left_index = True, right_index = True, how = 'left')
data = pd.merge(data, F3, left_index = True, right_index = True, how = 'left')
data = pd.merge(data, F4, left_index = True, right_index = True, how = 'left')
```

```python
data.drop(['income_type', 'edu_type', 'family_type', 'house_type'], axis = 1, inplace = True)
```

```python
data['income_total'] = data['income_total'].astype(int)
data['family_size'] = data['family_size'].astype(int)
data['begin_month'] = data['begin_month'].astype(int)
```
