### 2. 예보 등급, 예보 등급에 따라 개수 표시.

```python
f1 = df[df['PM2.5'] >= 76]
f2 = df[(df['PM2.5'] >= 36) & (df['PM2.5'] < 76)]
f3 = df[(df['PM2.5'] >= 16) & (df['PM2.5'] < 36)]
f4 = df[df['PM2.5'] < 16]
```

```python
import numpy as np

df['예보등급'] = np.nan
```

```python
df['예보등급'][f1.index] = '매우 나쁨'
df['예보등급'][f2.index] = '나쁨'
df['예보등급'][f3.index] = '보통'
df['예보등급'][f4.index] = '좋음'
```

```python
cnt1 = f1['PM2.5'].count()
cnt2 = f2['PM2.5'].count()
cnt3 = f3['PM2.5'].count()
cnt4 = f4['PM2.5'].count()
```
