### 학습에 가능하도록 수정.
- StationId
- Date
- AQI_Bucket
- 예보 등급 -> Target

```python
F1 = pd.get_dummies(df['StationId'], prefix = 'StationId')
F1
```

```python
del df['Date']
df
```

```python
F2 = pd.get_dummies(df['AQI_Bucket'], prefix = 'AQI_Bucket')
F2
```

```python
df['예보 등급'].astype('category')
```

```python
df.info()
```

```python
del df['StationId']
del df['AQI_Bucket']
```

```python
df = pd.merge(df, F1, left_index = True, right_index = True, how = 'left')
df = pd.merge(df, F2, left_index = True, right_index = True, how = 'left')
```

```python
df
```

```python
df = df.fillna(method = 'ffill')
```

```python
df = df.fillna(method = 'bfill') # 맨 앞에 있는 것 중 nan 있으면 이것을 채워주기 위해서
```

```python
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, confusion_matrix
from sklearn.preprocessing import StandardScaler # 일단은 사용하지 않기로.
```

```python
target = df['예보 등급']
df.drop('예보 등급', axis = 1, inplace = True)
```

```python
df
```

```python
target
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, target, test_size = 0.1, random_state = 42)
```

```python
forest = RandomForestClassifier()
```

```python
forest.fit(X_train, y_train)
```

```python
y_pred = forest.predict(X_test)
```

```python
cross_val_score(estimator = forest, X = X_test, y = y_test, cv = 5)
```

```python
confusion_matrix(y_true = y_test, y_pred = y_pred)
```

#### Random Grid Search로 성능 향상

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
forest_ = RandomForestClassifier()
distributions = {'max_depth' : [12, 24], 'n_estimators' : [100, 1000], 'max_leaf_nodes' : [8, 16, 24]}
clf = RandomizedSearchCV(forest_, distributions, random_state = 42)

search = clf.fit(X_test, y_test)
search.best_params_
```
