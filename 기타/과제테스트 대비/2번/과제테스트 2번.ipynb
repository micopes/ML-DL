```python
data = pd.read_csv('access.log', delimiter = ' ', header = None)
data
```

```python
del data[1]
del data[2]
del data[4]
data
```

```python
data[3] += ']'
data
```

```python
dic = {}
for i in range(len(data)):
    try:
        dic[data[0][i]] += 1
    except:
        dic[data[0][i]] = 1
```

```python
dic
```

```python
ans = pd.DataFrame(dic.items(), conclumns = ['IP', 'IP Address Count'])
ans
```

```python
ans.sort_values(by = 'IP Address Count', inplace = True)
ans
```
